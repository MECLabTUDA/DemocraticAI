```python
import pandas as pd
```

```python
df = pd.read_csv("../measure_model_results.csv")
```

```python
results = []
for index, row in df.iterrows():
    if row["train"] == "eval":  
        continue
    eval_row = df.iloc[index+1]
    assert row["data"] == eval_row["data"]
    assert row["model"] == eval_row["model"]
    results.append({
        "data": row["data"],
        "model": eval_row["model"],
        "params": eval_row["params"],
        "macs": eval_row["macs"],
        "flops_train": row["flops"],
        "mem_train": row["mem"],
        "time_train": row["time"],
        "energy_train": row["energy"],
        "flops_eval": eval_row["flops"],
        "mem_eval": eval_row["mem"],
        "time_eval": eval_row["time"],
        "energy_eval": eval_row["energy"],
    })
```

```python
results = pd.DataFrame(results)
model_map ={
    "unet": "UNet",
    "new_mednca": "FedNCA",
    "transunet_b16": "TransUNet",
    "segformer": "SegFormer",
    "vit": "ViT",
    "resnet18": "ResNet18",
    "maxmednca": "FedNCA"
}
results["model"] = results["model"].apply(lambda x: model_map.get(x, x))
results["params"] = results["params"].astype(int)
results["macs"] = results["macs"].apply(lambda x: x / 10**9)
results["flops_train"] = results["flops_train"].apply(lambda x: x / 10**9)
results["flops_eval"] = results["flops_eval"].apply(lambda x: x / 10**9)
results["mem_train"] = results["mem_train"].apply(lambda x: x / 10**6)
results["mem_eval"] = results["mem_eval"].apply(lambda x: x / 10**6)
```
